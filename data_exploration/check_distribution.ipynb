```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

```python
def plot_distrib(avg, var, shift=0, size=5, distrib=stats.norm.pdf, mult=1):
    sigma = math.sqrt(var)
    x = np.linspace(avg - size*sigma, avg + size*sigma, 100)
    plt.plot(mult*x, distrib(x - shift, avg, sigma))
    plt.show()
```

```python
def plot_triangular(left, mode, right): #needs to be corrected
    h = plt.hist(np.random.triangular(left, mode, right, 10000), bins=200)
    plt.show()
```

```python
# horse = max(0, np.random.normal(5,2,1)[0])
plot_distrib(avg=5, var=2, distrib = stats.norm.pdf)
```

```python
# ball = max(0, 1 + np.random.normal(1,0.3,1)[0])
plot_distrib(avg=1, var=0.3, shift=1)
```

```python
# bike = max(0, np.random.normal(20,10,1)[0])
plot_distrib(avg=20, var=10)
```

```python
# train = max(0, np.random.normal(10,5,1)[0])
plot_distrib(avg=10, var=5)
```

```python
# coal = 47 * np.random.beta(0.5,0.5,1)[0]
plot_distrib(avg=0.5, var=0.5, distrib=stats.beta.pdf, size=4, mult=47)
```

```python
# book = np.random.chisquare(2,1)[0]
plot_distrib(avg=2, var=1, distrib=stats.chi2.pdf,size=8)
```
