```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

```python
def plot_distrib(gift):
    mu, sigma = stats.norm.fit(gift)
    plt.hist(gift, bins=25, density=True, alpha=0.7, color='g')
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, mu, sigma)
    plt.plot(x, p, color='b', linewidth=2)
    title = "Fit results: mu = %.2f, sigma = %.2f" % (mu, sigma)
    plt.title(title)
    plt.show()
```

```python
size = 1000000

horse = np.maximum(0, np.random.normal(5,2,size))
ball = np.maximum(0, 1 + np.random.normal(1,0.3,size))
bike = np.maximum(0, np.random.normal(20,10,size))
train = np.maximum(0, np.random.normal(10,5,size))
coal = 47 * np.random.beta(0.5,0.5,size)
book = np.random.chisquare(2,size)
doll = np.random.gamma(5,1,size)
block = np.random.triangular(5,10,20,size)
gloves = np.random.rand(size);

for i in range(len(gloves)):
    if gloves[i] < 0.3:
        gloves[i] = 3 + np.random.rand(1)[0]
    else:
        gloves[i] = np.random.rand(1)[0]
```

```python
# Horse
plot_distrib(horse)
```

```python
# Ball
plot_distrib(ball)
```

```python
# Bike
plot_distrib(bike)
```

```python
# Train
plot_distrib(train)
```

```python
# Coal
plot_distrib(coal)
```

```python
# Book
plot_distrib(book)
```

```python
# Doll
plot_distrib(doll)
```

```python
# Block
plot_distrib(block)
```

```python
# Gloves
plot_distrib(gloves)
```
